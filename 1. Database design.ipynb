## **Database design**

In the last dictionary, we have explored the tables we want to integrate. Below, you can see a diagram of the tables.

![Diagram of current tables.drawio.png](<attachment:Diagram of current tables.drawio.png>)

Many of those columns contain information that is uninteresting or too incomplete to be worth it to include it.

![Diagram with deleted columns.png](<attachment:Diagram with deleted columns.png>)

Deeper analysis can be found in the EDA.ipynb document, but to make it short:

- STATUS_GZP, DESC_I, DESC_D (in both avalanches and landslides datasets) only have one distinct value. Every row says "approved". This is redundant, only approved rows should be included in the dataset anyway
- GEM_ID, in the same datasets, seemed to include information redundant with ISTAT_CODE, but of lower quality.
- ID_PHEN, in the avalanches dataset, is completely empty.
- All the columns discarded in the road segments dataset were extremely sparse and had no interesting information.

Ultimately, this was the chosen design.



Much of the information about roads was left out, as it would be out of the intended domain.
